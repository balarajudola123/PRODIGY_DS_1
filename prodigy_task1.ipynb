```python
# Check if any columns in the dataframe have missing values
missing_values = df.isna().sum().any()

if missing_values:
    # Print a warning message
    print("Warning: The dataframe contains missing values. This may cause errors when plotting histograms.")
```

```python
# Filter out rows with missing values
df_filtered = df.dropna()

# Plot histograms for the filtered dataframe
for i in col:
    fig=plt.figure(figsize=(3,3))
    plt.hist(df_filtered[i],bins=8)
    plt.xlabel(i)
    plt.show()
```
